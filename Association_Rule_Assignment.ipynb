```python

import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
item_list = pd.read_excel(r'C:\Users\swalp\OneDrive\Desktop\Data Analyst\DATA SCIENCE\new Assignment\Association Rules\Online retail.xlsx',header=None,names=["Transaction"])
item_list.head()
```

#### converting the values in transcation columns in to string.Process would be like
step 1) selects the column.

step 2) Converts each row into a string, splits it by commas, and returns a list of items for that transaction.

step 3) Turns the pandas Series of lists into a Python list of lists.

and finally showing only for 5 records for demo purpose by for loop.


```python
list = item_list["Transaction"].apply(lambda x: str(x).split(",")).tolist()
for t in list[:5]:
    print(t)
```

##### using Encoder to convert the values in numeric format, so that model understands it.

```python
te=TransactionEncoder()
data=te.fit_transform(list,sparse=False)
```

```python
df = pd.DataFrame(data, columns=te.columns_)
df
```

#### Creating frequent itemsets by using Apriori algo

```python
freq_items = apriori(df, min_support=0.01, use_colnames=True)
freq_items
## it can very upon choosing min_support differntly
```

```python
freq_items1 = apriori(df, min_support=0.005, use_colnames=True)
freq_items1
```

#### Extracting the association rule

```python
assoc_rules = association_rules(freq_items, min_threshold=0.5)
assoc_rules
# diifernt association  rules can be extracted by  changing  differnt min_threshold values)
```

```python
assoc_rules1 = association_rules(freq_items1, min_threshold=0.5)
assoc_rules1
```

#### final associated items based on highest lifts...

```python
assoc_rules.sort_values(by='lift', ascending=False).iloc[:3,:]
```

```python
assoc_rules1.sort_values(by='lift', ascending=False).iloc[:3,:]
```
